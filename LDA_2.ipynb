```python
from sklearn.datasets import fetch_20newsgroups
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
newsgroups=fetch_20newsgroups(subset="train",remove=("headers","footers","quotes"))
```

```python
documents=newsgroups.data
```

```python
print(f"Lodead {len(documents)} news group articals successfully !")
```

```python
print("printing the first 3 documents")
for i, doc in enumerate(documents[:3]):
    print(f"Document {i+1}:\n{doc}\n")
```

```python
vectorizer = CountVectorizer(stop_words='english')
word_counts = vectorizer.fit_transform(documents)
feature_names = vectorizer.get_feature_names_out()

lda = LatentDirichletAllocation(n_components=6, random_state=42)
lda.fit(word_counts)

print("Discovered Topics:- ")
for topic_idx, topic in enumerate(lda.components_):
  top_words_idx = topic.argsort()[:-20:-1]
  top_words = [feature_names[i] for i in top_words_idx]
  print(f"Hidden Topic #{topic_idx + 1}: {','.join(top_words)}")
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
word_counts = vectorizer.fit_transform(documents)
feature_names = vectorizer.get_feature_names_out()

lda = LatentDirichletAllocation(n_components=3, random_state=42)
lda.fit(word_counts)

print("Discovered Topics:- ")
for topic_idx, topic in enumerate(lda.components_):
  top_words_idx = topic.argsort()[:-7:-1]
  top_words = [feature_names[i] for i in top_words_idx]
  print(f"Hidden Topic #{topic_idx + 1}: {','.join(top_words)}")
```
